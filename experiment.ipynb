## Load the Data Set

```python
df = pd.read_csv("data/Churn_Modelling.csv")
```

## Preprocess the data
Drop irrelevant columns

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],index=1,inplace=True)
```

```python
df
```

```python
## Encode Categorical Variables
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
```

```python
df['Gender'].unique()
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder

onehot_encoder_geo = OneHotEncoder(sparse_output=False)
```

```python
geo_encode = onehot_encoder_geo.fit_transform(df[['Geography']])
```

```python
geo_encode
```

```python
geo_encoded_df=pd.DataFrame(geo_encode,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
## combine columns
```

```python
df = pd.concat(
    [df.drop('Geography', axis=1), geo_encoded_df],
    axis=1
)
```

```python
df
```

```python
df = df.dropna()
```
